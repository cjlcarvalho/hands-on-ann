# AND/OR Perceptron

```
weights = [0.0, 0.0, 0.0]
epochs = 30
data = [[0, 1, 0], [1, 0, 0], [0, 0, 0], [1, 1, 1]]
```

```
def activate(value):

    return 1 if value >= 0 else 0
```

```
# Perceptron main function
def perceptron(x1, x2):

    # Sum inputs to weights and bias
    net = (x1 * weights[0]) + (x2 * weights[1]) + ((-1) * weights[2])
    # Execute the activation function on the result
    return activate(net)
```

```
# Optimize weights in case of errors
def optimizeWeights(i, output):

    weights[0] += (data[i][2] - output) * data[i][0]
    weights[1] += (data[i][2] - output) * data[i][1]
    weights[2] += (data[i][2] - output) * -1 # bias
```

```
def training():

    count = 0

    while count < epochs:
        for i in range(len(data)):
            output = perceptron(data[i][0], data[i][1])
            # Optimize weights if the output is not equal to the expected value
            if output != data[i][2]:
                optimizeWeights(i, output)

        count += 1
```

```
def predict(sample):

    sample[2] = perceptron(sample[0], sample[1])
```

```
def print_weights():
    for i in weights:
        print(i, end=",")
    print()

print('Weights before training:')
print_weights()

training()

print('Weights after training:')
print_weights()
```

```
samples = [[0, 1, -1], [1, 0, -1], [0, 0, -1], [1, 1, -1]]

for sample in samples:
    predict(sample)

print('Result:')

for sample in samples:
    for s in sample:
        print(s, end=',')
    print()
```
